```python
# Changing the data types to datetime for 'Date Rptd'
LA['Date Rptd'] = pd.to_datetime(LA['Date Rptd'])
```

```python
# Check if it worked
LA.info()
```

```python
# Create a datetime with 'DATE OCC' and 'TIME OCC'

# Convert 'DATE OCC' to string
LA['DATE OCC'] = LA['DATE OCC'].astype(str)

# Convert 'TIME OCC' to string and format it to have ':' before the last two digits
LA['TIME OCC'] = LA['TIME OCC'].apply(lambda x: str(x).zfill(4)) # Pad with zeros to have 4 digits
LA['TIME OCC'] = LA['TIME OCC'].apply(lambda x: x[:-2] + ':' + x[-2:]) # Insert ':' before the last two digits

# Combine 'DATE OCC' and 'TIME OCC' into a single column
LA['datetime_OCC'] = pd.to_datetime(LA['DATE OCC'] + ' ' + LA['TIME OCC'])
```

```python
# Check if it worked
LA.info()
```

```python
# Drop old columns
LA = LA.drop(columns=['DATE OCC', 'TIME OCC'])
```

```python
# Drop unnecessary columns
LA = LA.drop(columns=['Crm Cd', 'Premis Cd', 'Crm Cd 1','Crm Cd 2', 'Weapon Used Cd', 'Status', 'Status Desc', 'Crm Cd 3', 'Crm Cd 4', 'LOCATION', 'Cross Street', 'AREA NAME', 'Rpt Dist No', 'Part 1-2', 'Mocodes', 'Date Rptd'])
```
